#### 5. Convert datatype of column published_at from object to pandas datetime.

```python
data1['published_at']=pd.to_datetime(data1['published_at'])
```

```python
data1.info()
```

#### 6. Create a new column as 'published_month' using the column published_at (display the months only)


```python
data1['published_month']=data1['published_at'].dt.month
```

```python
data1.info()
```

#### 7. Replace the numbers in the column published_month as names of the months i,e., 1 as 'Jan', 2 as 'Feb' and so on.....


```python
# extracting the month name from 'published_at' column and saving it in published_month column
data1['published_month']=data1['published_at'].dt.month_name()
# Slicing the first three letters of each month name and saving in the published_month column
data1['published_month'] = data1['published_month'].str[0:3]
```

```python
data1
```

#### 8. Find the number of videos published each month and arrange the months in a decreasing order based on the video count.

```python
# 1. Grouping the dataset for each month.
# 2. Counting the number of videos in each month
# 3. Sorting the obtained data into ascending order based on number of videos count
data1.groupby('published_month')['video_id'].count().sort_values(ascending=False)
```

#### 9. Find the count of unique video_id, channel_id and channel_title.

#### 10. Find the top10 channel names having the highest number of videos in the dataset and the bottom10 having lowest number of videos.


```python
# 1. Grouping the dataset for each channel_title.
# 2. Counting the number of videos in each channel_title
# 3. Sorting the obtained data into ascending order based on number of videos count and printing for first 10 channels
data1.groupby('channel_title')['video_id'].count().sort_values(ascending=False).head(10)
```

```python
# 1. Grouping the dataset for each channel_title.
# 2. Counting the number of videos in each channel_title
# 3. Sorting the obtained data into ascending order based on number of videos count and printing for last 10 channels
data1.groupby('channel_title')['video_id'].count().sort_values(ascending=False).tail(10)
```

#### 13. Does the number of views have any effect on how many people disliked the video? Support your answer with a metric and a plot.


```python
plt.figure(figsize=(10,5))
plt.scatter(data1['view_count'],data1['dislikes'])
plt.xlabel('view_count')
plt.ylabel('dislikes')
```

```python
data1.corr()
```

From the above metric, we get to know that, correlation between views count and number of dislikes is 0.684, which is not a bad figure as compared to real world. Therefore, it can be concluded that number of views count have significant impact on number of dislikes.
